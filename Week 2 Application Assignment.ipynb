### Model 2: only the target variable is transformed

```python
y = df['logmonth']
x1 = df['NumStores']
```

```python
x = sm.add_constant(x1)
```

```python
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

### Model 3: only the explanatory variable is transformed

```python
results = sm.OLS(y,x).fit()
```
